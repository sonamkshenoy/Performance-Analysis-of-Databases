```python
import csv
import matplotlib.pyplot as plt
import numpy as np

filename = 'newcleaned.csv'
headers=[]
dbs=[]
data=[]
read1=[]
read2=[]
read3=[]
read4=[]
read5=[]
read6=[]
write1=[]
write2=[]
write3=[]
write4=[]
write5=[]
write6=[]

with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    for index,column_header in enumerate(header_row):
        #print(index, column_header)
        if(column_header!=''):
            headers.append(column_header)
    i=1
    for row in reader:
        dbs.append(row[0])
        read1.append(float(row[3]))
        read2.append(float(row[4]))
        read3.append(float(row[5]))
        read4.append(float(row[6]))
        read5.append(float(row[7]))
        read6.append(float(row[8]))
        write1.append(float(row[16]))
        write2.append(float(row[17]))
        write3.append(float(row[18]))
        write4.append(float(row[19]))
        write5.append(float(row[20]))
        write6.append(float(row[21]))

# print(headers)
# print(dbs)
# print(read1)
# print(read2)
# print(read3)
# print(read4)
# print(read5)
# print(read6)
# print(read7)
# headers_proper=[]
# for i in headers:
#     if(i[1:5]=="READ"):
#         headers_proper.append(i[7:])
#     elif(i[1:7]=="UPDATE"):
#         headers_proper.append(i[9:])
#     else:
#         headers_proper.append(i)
        
# print(headers_proper)
print(plt.style.available)
plt.style.use('seaborn-whitegrid')
# plt.style.available
width = 0.07
y_pos = [0,0.4]

plt.bar(y_pos, read1,width=0.07,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("No. of Operations")
plt.xlabel("Database")
plt.title("Comparison of No. of Operations")
plt.show()

plt.bar(y_pos, read2,width=0.07,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("Latency(s)")
plt.xlabel("Database")
plt.title("Comparison of Average Latency")
plt.show()

plt.bar(y_pos, read3, width=width,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("Latency(s)")
plt.xlabel("Database")
plt.title("Comparison of Minimum Latency")
plt.show()

plt.bar(y_pos, read4, width=width,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("Latency(s)")
plt.xlabel("Database")
plt.title("Comparison of Maximum Latency")
plt.show()

plt.bar(y_pos, read5, width=width,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("Latency(s)")
plt.xlabel("Database")
plt.title("Comparison of 95 Percentile Latency")
plt.show()

plt.bar(y_pos, read6, width=width,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("Latency(s)")
plt.xlabel("Database")
plt.title("Comparison of 99 Percentile Latency")
plt.show()

#Summary
barWidth = 0.25
 
bars1 = [read2[0],read3[0],read4[0],read5[0],read6[0]]
bars2 = [read2[1],read3[1],read4[1],read5[1],read6[1]]

r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
 
plt.bar(r1, bars1, color='lightblue', width=barWidth, edgecolor='white', label='Couchbase')
plt.bar(r2, bars2, width=barWidth, edgecolor='white', label='MongoDB')
plt.xlabel('Database', fontweight='bold')
plt.ylabel("Latency (ms)", fontweight='bold')
plt.xticks([r-0.15 + barWidth for r in range(len(bars1))], ['Average Latency', 'Min Latency', 'Max Latency','95 Percentile','99 Percentile'])
plt.title("Comparison of Read Latency")
 

plt.legend()
plt.show()
###########################################################################################################
plt.bar(y_pos, write1,width=0.07,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("No. of Operations")
plt.xlabel("Database")
plt.title("Comparison of No. of Operations")
plt.show()

plt.bar(y_pos, write2,width=0.07,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("Latency(s)")
plt.xlabel("Database")
plt.title("Comparison of Average Latency")
plt.show()

plt.bar(y_pos, write3, width=width,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("Latency(s)")
plt.xlabel("Database")
plt.title("Comparison of Minimum Latency")
plt.show()

plt.bar(y_pos, write4, width=width,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("Latency(s)")
plt.xlabel("Database")
plt.title("Comparison of Maximum Latency")
plt.show()

plt.bar(y_pos, write5, width=width,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("Latency(s)")
plt.xlabel("Database")
plt.title("Comparison of 95 Percentile Latency")
plt.show()

plt.bar(y_pos, write6, width=width,edgecolor='black')
plt.xticks(y_pos, dbs)
plt.ylabel("Latency(s)")
plt.xlabel("Database")
plt.title("Comparison of 99 Percentile Latency", fontweight='bold')
plt.show()

#Summary
barWidth = 0.25
 
bars1 = [write2[0],write3[0],write4[0],write5[0],write6[0]]
bars2 = [write2[1],write3[1],write4[1],write5[1],write6[1]]

r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
 
plt.bar(r1, bars1, color='lightblue', width=barWidth, edgecolor='white', label='Couchbase')
plt.bar(r2, bars2, width=barWidth, edgecolor='white', label='MongoDB')
plt.xlabel('Database', fontweight='bold')
plt.ylabel("Latency (s)", fontweight='bold')
plt.xticks([r-0.15 + barWidth for r in range(len(bars1))], ['Average Latency', 'Min Latency', 'Max Latency','95 Percentile','99 Percentile'])
plt.title("Comparison of Write Latency")
 

plt.legend()
plt.show()
    
```
